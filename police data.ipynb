# OUTLIER REMOVAL

```python
import seaborn as sns
sns.boxplot(x=data['driver_age'])   # scatter plot for knowing the outliers
```

```python
# driver age has outliers, so we have to solve the issue of the outliers, also driver age has many null values.
# first we will replace all the null values in driver age by driver's age median.
```

```python
data["driver_age"].fillna(data['driver_age'].median(), inplace = True)
```

```python
data['driver_age'].isnull().sum()
```

```python
# we are using z score for handling outliers

```

```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(data['driver_age']))
print(z)
```

```python
data['zcore_of_drivers_age']=pd.DataFrame(z)
```

```python
data
```

```python
# now before removing the outliers we have 65535 rows.... 
```

```python
# removing thr outliers
```

```python
data=data[(data.zcore_of_drivers_age>-3) &(data.zcore_of_drivers_age<3)]
```

```python
data
```

```python
# drawing boxplot after removing the outliers, before removing the unnecessary outliers , we had 65535 rows, but now we have 65245 rows.
sns.boxplot(x=data['driver_age'])
```

# USING IQR FOR OUTLIER DETECTION

```python
iqr_data

sns.boxplot(x=iqr_data['driver_age'])
```

# IQR for detecting outliers

```python
Q1 = iqr_data.driver_age.quantile(0.25)
Q3 = iqr_data.driver_age.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
iqr_data[(iqr_data.driver_age<lower_limit)|(iqr_data.driver_age>upper_limit)]
```

```python
iqr_data_no_outlier = iqr_data[(iqr_data.driver_age>lower_limit)&(iqr_data.driver_age<upper_limit)]
iqr_data_no_outlier
```

```python
data.isnull().sum()


```

```python
# we can see all are features that are having missing values are categorical variables and they are approx 4 thousand 
# rows , so deleting 4k rows would result in a lot of information loss, so we will go with simple imputer.
```

```python
data['stop_duration']=data['stop_duration'].fillna('0-15 Min')
data['is_arrested']=data['is_arrested'].fillna('False')
data['stop_outcome']=data['stop_outcome'].fillna('Citation')
data['violation']=data['violation'].fillna('Speeding')
data['driver_gender']=data['driver_gender'].fillna('M')
```
